```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Figureと3DAxeS
fig = plt.figure(figsize = (12, 12))
ax = fig.add_subplot(111, projection="3d")

# 軸ラベルを設定
# ax.set_xlabel("x", size = 16)
# ax.set_ylabel("y", size = 16)
# ax.set_zlabel("z", size = 16)

# 円周率の定義
pi = np.pi

# (x,y)データを作成
x = np.linspace(-5*pi, 5*pi, 2024)
y = np.linspace(-5*pi, 5*pi, 2024)

# 格子点を作成
X, Y = np.meshgrid(x, y)

# 高度の計算式
Z = np.cos(X/pi*0.4) * np.sin(Y/pi*0.5)

# 曲面を描画
ax.plot_surface(X, Y, Z, shade=False)
# ax.plot_wireframe(X, Y, Z, hide=True)

# hide grid
ax.grid(False)

# Hide axes ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

# 底面に等高線を描画
# ax.contour(X, Y, Z, colors = "black", offset = -1)

# fig.savefig("./img.png")
plt.axis('off')
plt.show()
```
